```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

#**Pre-Processing**#

```python
#import dataset
assessment = pd.read_csv("/content/assessments.csv")
assessment.head()
```

```python
#mengecek missing value
print(assessment.isnull().values.any())
```

```python
#drop missing value
assessment.dropna(axis=0, inplace=True)
```

```python
#mengecek data duplikat
assessment.duplicated().values.any()
```

```python
#Mengambil nilai unik
assessment['code_module'].unique()
```

```python
assessment['assessment_type'].unique()
```

```python
assessment.describe()
```

```python
assessment.info()
```

```python
labelencoder = LabelEncoder()

assessment['code_module'] = labelencoder.fit_transform(assessment['code_module'])
assessment['code_presentation'] = labelencoder.fit_transform(assessment['code_presentation'])
assessment['assessment_type'] = labelencoder.fit_transform(assessment['assessment_type'])
assessment['date'] = labelencoder.fit_transform(assessment['date'])
assessment['weight'] = labelencoder.fit_transform(assessment['weight'])
```
